```python
import pandas as pd
import pyreadstat
import os

# Load the dataset with metadata
file_path = "./CAUCASUS BAROMETER/cb2019_CB_2019_Regional_25.06.2020.sav"
df, meta = pyreadstat.read_sav(file_path, encoding='UTF-8', disable_datetime_conversion=True)

# List of variables to check
variables_to_check = [
    'RESPAGE', 'LIFESAT', 'INFOSOU1', 'TRUMTVINFO', 'FATEINLF', 'BUSINRUS', 'BUSINTUR', 
    'BUSINGA', 'NEIGHBOR', 'USSRDISS', 'EUSUPP', 'EEUSUPNA', 'EEUNSUW', 'EEUSUPW', 
    'MAINFRN', 'MAINENEM', 'MAINMOD', 'PARTYSUPP', 'PARTYSUPF', 'WORKTYP', 'PERSINC', 
    'EDUYRS', 'RFAEDUC', 'RMOEDUC', 'RELGNEW', 'RELIMP', 'ECONSTN', 'MONYTOT'
]

# Prepare a list to hold the data
output_data = []

# Extract the variable names, descriptions, and value labels
for variable in variables_to_check:
    if variable in meta.column_names:
        # Get the label (description) for the variable
        label = meta.column_labels[meta.column_names.index(variable)]
        
        # Get the value labels (scoring system) if available
        if variable in meta.variable_to_label:
            value_labels = meta.value_labels.get(meta.variable_to_label[variable], {})
            if value_labels:
                for value, vlabel in value_labels.items():
                    output_data.append([variable, label, value, vlabel])
            else:
                output_data.append([variable, label, None, None])
        else:
            output_data.append([variable, label, None, None])
    else:
        output_data.append([variable, "Label not found", None, None])

# Convert to DataFrame
output_df = pd.DataFrame(output_data, columns=['Variable', 'Description', 'Value', 'Value Label'])

# Ensure the directory exists
output_dir = "./data/"
os.makedirs(output_dir, exist_ok=True)

# Save to CSV
output_csv_path = os.path.join(output_dir, "variable_labels_and_values.csv")
output_df.to_csv(output_csv_path, index=False)

print(f"Variable labels and value labels saved to: {output_csv_path}")
```

```python
import pandas as pd
import pyreadstat
import numpy as np
import os

# Load the dataset with metadata
file_path = "./CAUCASUS BAROMETER/cb2019_CB_2019_Regional_25.06.2020.sav"
df, meta = pyreadstat.read_sav(file_path, encoding='UTF-8', disable_datetime_conversion=True)

# Replace values less than 0 with NaN to ignore them in correlation calculations
df_cleaned = df.applymap(lambda x: x if x >= 0 else np.nan)

# Compute the correlation matrix
corr_matrix = df_cleaned.corr(method='pearson', min_periods=1)

# Filter correlations above 0.7 or below -0.7
filtered_corr = corr_matrix[(corr_matrix.abs() > 0.7) & (corr_matrix != 1.0)].stack().reset_index()
filtered_corr.columns = ['Variable 1', 'Variable 2', 'Correlation']

# Add labels (descriptions) to the variables
filtered_corr['Label 1'] = filtered_corr['Variable 1'].map(dict(zip(meta.column_names, meta.column_labels)))
filtered_corr['Label 2'] = filtered_corr['Variable 2'].map(dict(zip(meta.column_names, meta.column_labels)))

# Rearrange columns to include labels next to variable names
filtered_corr = filtered_corr[['Variable 1', 'Label 1', 'Variable 2', 'Label 2', 'Correlation']]

# Ensure the directory exists
output_dir = "./data/"
os.makedirs(output_dir, exist_ok=True)

# Save the filtered correlations to a CSV file
output_csv_path = os.path.join(output_dir, "filtered_correlations_with_labels.csv")
filtered_corr.to_csv(output_csv_path, index=False)

print(f"Filtered correlations with labels saved to: {output_csv_path}")
```

```python
import pandas as pd
import pyreadstat

# Load the dataset with metadata
file_path = "./CAUCASUS BAROMETER/cb2019_CB_2019_Regional_25.06.2020.sav"
df, meta = pyreadstat.read_sav(file_path, encoding='UTF-8', disable_datetime_conversion=True)

# List of variables to check
variables_to_check = [
    'USSRDISS','RESPSEX'
]

# Print the variable names, descriptions, and their value labels
for variable in variables_to_check:
    if variable in meta.column_names:
        # Get the label (description) for the variable
        label = meta.column_labels[meta.column_names.index(variable)]
        print(f"{variable}: {label}")
        
        # Get the value labels (scoring system) if available
        if variable in meta.variable_to_label:
            value_labels = meta.value_labels.get(meta.variable_to_label[variable], {})
            if value_labels:
                print("  Value Labels (Scoring System):")
                for value, label in value_labels.items():
                    print(f"    {value}: {label}")
        else:
            print("  No value labels available for this variable.")
    else:
        print(f"{variable}: Label not found")
    print()  # Add space between variables
```

```python
import pandas as pd
import pyreadstat
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "./CAUCASUS BAROMETER/cb2019_CB_2019_Regional_25.06.2020.sav"
df, meta = pyreadstat.read_sav(file_path, encoding='UTF-8', disable_datetime_conversion=True)

# Filter the dataset to include only relevant responses for REPSEX and USSRDISS
df_filtered = df[(df['RESPSEX'].isin([1.0, 2.0])) & (df['USSRDISS'].isin([1.0, 2.0]))]

# Map the numeric values to their corresponding labels for better readability in the plot
df_filtered['RESPSEX'] = df_filtered['RESPSEX'].map({1.0: 'Male', 2.0: 'Female'})
df_filtered['USSRDISS'] = df_filtered['USSRDISS'].map({1.0: 'A good thing', 2.0: 'A bad thing'})

# Create a count plot to visualize the relationship between RESPSEX and USSRDISS
plt.figure(figsize=(10, 6))
sns.countplot(data=df_filtered, x='USSRDISS', hue='RESPSEX')
plt.title('Opinions on the Dissolution of the Soviet Union by Gender')
plt.xlabel('Opinion on USSR Dissolution')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```
